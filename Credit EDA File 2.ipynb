# DATA IMPUTATION - MISSING VALUES.

```python
print((df.isnull().mean()*100).sort_values(ascending=False))  # To find the percentage of null values in the data set
```

```python
# Dropping the columns having null values more than 40%

perc = 40.0
min_count =  int(((100-perc)/100)*df.shape[0] + 1)
df1 = df.dropna( axis=1, thresh=min_count)
```

```python
cont_cols=[df1["AMT_GOODS_PRICE"],df1["AMT_ANNUITY"],df1["CNT_PAYMENT"]]
```

```python
# Finding the normal distribution to IMPUTE the values using mean/median/mode.

plt.figure(figsize=(26,24))
cnt=1
for i in cont_cols:
    plt.subplot(1,3,cnt)
    sns.distplot(i,hist=False)
    cnt+=1
    i+=1

plt.show()
    
```

# As per the graph suggests the normal distribution level of each columns, and imputing its null values as below:--

```python
df1["AMT_GOODS_PRICE"].fillna(df1["AMT_GOODS_PRICE"].median(),inplace=True)
```

```python
df1["AMT_ANNUITY"].fillna(df1["AMT_ANNUITY"].median(),inplace=True)
```

```python
df1["CNT_PAYMENT"].fillna(df1["CNT_PAYMENT"].median(),inplace=True)
```

```python
df1["PRODUCT_COMBINATION"].fillna(df1["PRODUCT_COMBINATION"].mode(),inplace=True)
```
