# DATA IMPUTATION-MISSING VALUES

```python
print((df.isnull().mean()*100).sort_values(ascending=False))     # To find the percentage of null values in the data set
```

```python
# Dropping the columns having null values more than 40%

perc = 40.0
min_count =  int(((100-perc)/100)*df.shape[0] + 1)
df1 = df.dropna( axis=1, thresh=min_count)
```

```python
# Saving the columns with >40 % null values

cont_cols=[df['AMT_REQ_CREDIT_BUREAU_HOUR'],df["AMT_REQ_CREDIT_BUREAU_DAY"],df["AMT_REQ_CREDIT_BUREAU_WEEK"],
           df["AMT_REQ_CREDIT_BUREAU_MON"],df["AMT_REQ_CREDIT_BUREAU_QRT"],df["AMT_REQ_CREDIT_BUREAU_YEAR"],
           df["AMT_ANNUITY"],df["AMT_GOODS_PRICE"],df["DAYS_LAST_PHONE_CHANGE"],df["CNT_FAM_MEMBERS"],
           df["OBS_30_CNT_SOCIAL_CIRCLE"],df["DEF_30_CNT_SOCIAL_CIRCLE"]]
```

```python
# Finding the normal distribution to IMPUTE the values using mean/median/mode.

plt.figure(figsize=(26,24))
cnt=1
for i in cont_cols:
    plt.subplot(4,3,cnt)
    sns.distplot(i,hist=False)
    cnt+=1
    i+=1

plt.show()
    
```

# As per the graph suggests the normal distribution level of each columns, and imputing its null values as below:--

```python
df1["AMT_REQ_CREDIT_BUREAU_HOUR"].fillna((df1["AMT_REQ_CREDIT_BUREAU_HOUR"].median()),inplace=True)
```

```python
df1["AMT_REQ_CREDIT_BUREAU_DAY"].fillna((df1["AMT_REQ_CREDIT_BUREAU_DAY"].median()),inplace=True)
```
